```python
import numpy as np
import tensorflow as tf
import cv2
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, Flatten, BatchNormalization, Conv2D, MaxPool2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import itertools
import os
import shutil
import random
import glob
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
%matplotlib inline
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print("Num GPUs Available: ", len(physical_devices))
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
trainFile = 'dataset2/train/'
validFile = 'dataset2/valid/'
testFile = 'dataset2/test/'
modelFile = 'modellen/'
#rgoot getal vanwege early stopping
epochs = 2000
batch_size = 32
img_size = 112
trait = "D2_zoom50"
info = "1515 images per bord, met rescale, filters 32>64, zoom 0.50"
```

```python
train_datagen = ImageDataGenerator(
        rescale=1 / 255.0,
        zoom_range=0.50,
        validation_split=0.20
)

train_batches = train_datagen.flow_from_directory(directory=trainFile, target_size=(img_size,img_size), classes=['100','50','30'], batch_size=batch_size, subset='training')
valid_batches = train_datagen.flow_from_directory(directory=trainFile, target_size=(img_size,img_size),classes=['100','50','30'], batch_size=batch_size,  subset='validation')
```

```python
print(train_batches.n,valid_batches.n)
assert train_batches.n == 3636
assert valid_batches.n == 909
assert train_batches.num_classes == valid_batches.num_classes == 3
```

```python
imgs, labels = next(train_batches)

fig, axes = plt.subplots(3, 3, figsize=(20,20))
axes = axes.flatten()
for img, ax in zip( imgs, axes):
        #img = img/255
        ax.imshow(img)
        ax.axis('off')
plt.tight_layout()
plt.show()

```

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=(3, 3), activation='relu', padding = 'same', input_shape=(img_size,img_size,3)),
    MaxPool2D(pool_size=(2, 2), strides=2),
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding = 'same'),
    MaxPool2D(pool_size=(2, 2), strides=2),
    Flatten(),
    #Dense(units=512, activation='relu'),
    #Dropout(0.8),
    Dense(units=3, activation='softmax')
])
```

```python
model.summary()
```

```python
#import os.path
modelNum = 1
savedModel = False
model_path = "."
for file in os.listdir(modelFile):
    print(file)
    if ".h5" in file:
        modelNum += 1
        
model_name = "Model"+str(modelNum)+"_Batch"+str(batch_size)+"_"+trait


if os.path.isfile(modelFile+model_name):
    model_name += "_copy"
    
```

```python
checkpoint = ModelCheckpoint(modelFile+model_name+".h5", 
                             monitor='val_accuracy', 
                             verbose=1, 
                             save_best_only=True, 
                             save_weights_only=False, 
                             mode='auto', 
                             save_freq="epoch")

early = EarlyStopping(monitor='val_accuracy', 
                      min_delta=0, patience=20, 
                      verbose=1, 
                      mode='auto',)
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x=train_batches,
    steps_per_epoch=len(train_batches),
    validation_data=valid_batches,
    validation_steps=len(valid_batches),
    epochs=epochs,
    verbose=2,
    callbacks=[checkpoint,early],
)
```

```python
#bewaar gegevens van beste val_accuracy
vAcc = round(max(history.history['val_accuracy']),2)
index = history.history['val_accuracy'].index(max(history.history['val_accuracy']))
tAcc = str(round(history.history['accuracy'][index], 2))
vLoss = str(round(history.history['val_loss'][index], 2))
tLoss = str(round(history.history['loss'][index], 2))

print("Val_accuracy: {}\nval_loss: {}\ntest_accuracy: {}\ntest_loss: {}\nbatch size: {}\nimg size: {}".format(vAcc,vLoss,tAcc,tLoss,batch_size,img_size))

# maakt file aan met model info
f = open(modelFile+model_name+".txt","w")
model_info = "Val_accuracy: {}\nVal_loss: {}\nTrain_accuracy: {}\nTrain_loss: {}\nBatch size: {}\nImg size: {}\ntraining data: {}\nvalid data: {}\nExtra info: {}".format(vAcc,vLoss,tAcc,tLoss,batch_size,img_size,train_batches.n,valid_batches.n,info)
f.write(model_info)
f.close()
```
